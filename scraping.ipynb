```python
from google_play_scraper import Sort, reviews
import pandas as pd
import time
```

```python
# Fungsi untuk scrape review dalam batch
def scrape_reviews(app_id, total_reviews=20000):
    count = 0
    result_list = []
    next_token = None
    
    while count < total_reviews:
        n_to_fetch = min(200, total_reviews - count)
        reviews_batch, next_token = reviews(
            app_id,
            lang='id',
            country='id',
            sort=Sort.NEWEST,
            count=n_to_fetch,
            continuation_token=next_token
        )
        result_list.extend(reviews_batch)
        count += len(reviews_batch)
        print(f'Scraped: {count} reviews')
        time.sleep(1)
        
        if next_token is None:
            break
        
    return result_list

# ID Aplikasi Shopee
app_id = 'com.shopee.id'

# Scrape 10.000 Review
data = scrape_reviews(app_id, total_reviews=20000)

#Simpan ke CSV
df = pd.DataFrame(data)[['userName', 'content', 'score', 'at']]
df.to_csv('shopee_reviews.csv', index=False)
print("Sudah disimpan ke shopee_reviews.csv")
```
